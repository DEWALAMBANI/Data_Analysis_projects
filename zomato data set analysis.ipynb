```python
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis = 1)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop_duplicates(inplace = True)
df.head()
```

```python
df.shape
```

```python
#cleaning rate column
df['rate'].unique()
```

```python
#removing new and - and /5 from rate column

def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value= value[0]
        return float(value)
    

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#filling null values in rate column with mean
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace= True)
```

```python
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'cost_of_2','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df.head()
```

```python
df['listed_in(city)'].unique()
```

```python
df.drop(['listed_in(city)'], axis=1)
```

```python
df['cost_of_2'].unique()
```

```python
#removing comma from no.s in between 
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
        
df['cost_of_2'] = df['cost_of_2'].apply(handlecomma)
df['cost_of_2'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = True)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types < 1000]
rest_types_lessthan1000
```

```python
#making less than 1000 in frequency as others

def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = True)
location
```

```python
location_lessthan300 = location[location < 300]
location_lessthan300
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
    
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
df
```

```python
df.drop(['listed_in(city)'], axis=1)
```

```python
df.head()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)
```

```python
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines < 100]
```

```python
cuisines_lessthan100
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
